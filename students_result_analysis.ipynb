```python
import pandas as pd
df = pd.read_excel("new result.xlsx")
df
```

```python
print(df.shape)
df.head()
```

```python
df[["Math","English","Computer Science"]].mean()
                                                            # average of all col
```

```python
df[["Math","English","Computer Science"]].max()
                                                             # max num of all col
```

```python
df[["Math","English","Computer Science"]].min()
                                                            # min num of all col
```

```python
df["Total"]=df[["Math","English","Computer Science"]].sum(axis=1)
df.head()
```

```python
subject_averages = df[["Math","English","Computer Science"]].mean().reset_index()
subject_averages.columns = ["Subject", "Average"]
print(subject_averages)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Subject", y="Average", data=subject_averages, palette="pastel")
plt.title("Average Marks per Subject")
plt.ylabel("Average Marks")
plt.ylim(0, 100)
plt.show()
```
